```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.load('saved_npy/cartoon.npy').mean(0),label='CartoonX')
plt.plot(np.load('saved_npy/randompixel.npy').mean(0),label='Random Pixel')
plt.plot(np.load('saved_npy/SmoothG.npy').mean(0),label='Smooth Grad')
plt.plot(np.load('saved_npy/randomwavelet.npy').mean(0),label='Random Wavelet')
plt.plot(np.load('saved_npy/IntegG.npy').mean(0),label='Integrated Gradient')
plt.plot(np.load('saved_npy/pixelrde.npy').mean(0),label='Pixel RDE')
plt.plot(np.load('saved_npy/GuidedBP.npy').mean(0),label='Guided Backprop')
plt.xlabel('Non-randomized relevant components')
plt.ylabel('Distortion')
plt.legend()
# plt.savefig('7a.png')
plt.show()
```
